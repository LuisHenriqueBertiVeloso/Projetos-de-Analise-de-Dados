```python
tabela = pd.read_csv('clientes.csv')
```

```python
display(tabela)
display(tabela.info())
```

```python
#Preparando a base de dados
#Precisamos deixar setado apenas uma coluna em texto que vai ser o objeto previsto
#As colunas profissao, mix_credito e comportamento_pagamento precisando viar codificadores
from sklearn.preprocessing import LabelEncoder

codificador = LabelEncoder()
tabela['profissao'] = codificador.fit_transform(tabela['profissao'])
tabela['mix_credito'] = codificador.fit_transform(tabela['mix_credito'])
tabela['comportamento_pagamento'] = codificador.fit_transform(tabela['comportamento_pagamento'])
tabela.info()
```

```python
#Um AI precisa de dados de aprendizado e dados de teste. Precisamos separar nossa base de dados
#nessas duas categorias
#depois separar nossos dados em x e y (y será a coluna dependente -o que se irá prever)
# o x (serão os dados usados para fazer a previsão)

y = tabela['score_credito']

colunas = ['id_cliente', 'score_credito']

x = tabela.drop(columns=colunas)


from sklearn.model_selection import train_test_split

x_treino, x_teste, y_treino, y_teste = train_test_split(x,y)
```

```python
#Agora vamos criar dois modelos de AI

#modelo árvore de decisão
from sklearn.ensemble import RandomForestClassifier

#Modelo KNN
from sklearn.neighbors import KNeighborsClassifier

modelo_arvoredecisao = RandomForestClassifier()
modelo_knn = KNeighborsClassifier()

modelo_arvoredecisao.fit(x_treino, y_treino)
modelo_knn.fit(x_treino, y_treino)
```

```python
#escolhendo o melhor modelo
from sklearn.metrics import accuracy_score #vai medir quantos acertos cada modelo teve

previsao_arvore = modelo_arvoredecisao.predict(x_teste)

previsao_knn = modelo_knn.predict(x_teste.to_numpy()) #precisa transformar seu x_teste de pandas pra NUMPY para o KNN

#vendo a accuracy

a = accuracy_score(y_teste, previsao_arvore)
b = accuracy_score(y_teste, previsao_knn)

print(a)
print(b)
```

```python
#Modelo de Arvore de Decisão teve uma melhor performance
```

```python
#Agora vamos fazer novas previsões

novos_clientes = pd.read_csv('novos_clientes.csv')
novos_clientes['profissao'] = codificador.fit_transform(novos_clientes['profissao'])
novos_clientes['mix_credito'] = codificador.fit_transform(novos_clientes['mix_credito'])
novos_clientes['comportamento_pagamento'] = codificador.fit_transform(novos_clientes['comportamento_pagamento'])

previsoes = modelo_arvoredecisao.predict(novos_clientes)
display(novos_clientes)
display(previsoes)
```
